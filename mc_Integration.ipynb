### Monte Carlo integration

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
x_range = np.linspace(0, 2, 1000)
x_sample = np.random.choice(x_range, size=1000)
def function(array):
    return (array ** 3) - (array ** 2)
y_sample = function(x_sample)
y_sample.mean() * 2
```

```python
plt.scatter(x_range, function(x_range), s=.5)
plt.title('Exact Values')
plt.show()
```

```python
plt.scatter(x_sample, y_sample, s=.5)
plt.title('Sample Values')
plt.show()
```
